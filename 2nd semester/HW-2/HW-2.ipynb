# kNN своими руками (5)

В этом задании вам предстоит реализовать взвешенный алгоритм kNN для регрессии. Пусть необходимо вычислить значение $y$ для некоторого $x$ при известных данных $\left(x_1, y_1\right), \ldots, \left(x_\ell, y_\ell\right)$. Предсказанием вашего регрессора будет являться

$$
\hat{y} = \frac{\sum\limits_{i=1}^kw_iy_{(i)}}{\sum\limits_{i=1}^kw_i},
$$
где $\left(x_{(1)}, y_{(1)}\right), \ldots, \left(x_{(k)}, y_{(k)}\right)$ - ближайшие $k$ объектов к $x$ по некоторой метрике $d(\cdot, \cdot)$. Ваш алгоритм должен уметь работать с двумя метриками:

$$
d\left(x_{(i)}, x\right) = \|x_{(i)} - x\|_2 = \sqrt{\sum\limits_{j=1}^n\left(x_{(i)}^j - x^j\right)^2}\qquad\text{(евклидова)}
$$
$$
d\left(x_{(i)}, x\right) = \|x_{(i)} - x\|_1 = \sum\limits_{j=1}^n\left|x_{(i)}^j - x^j\right|\qquad\text{(манхэттена)}
$$

### Реализуйте две функции расстояния (1 балл)
- евклидова метрика **(0.5 балла)**
- метрика Манхэттена **(0.5 балла)**

Обе функции должны на вход получать матрицу `np.array of shape(n, m)` и вектор `np.array of shape(m,)`, а возвращать вектор расстояний от каждой строчки матрицы до вектора `np.array of shape(n,)`

**В данном пункте запрещено использование циклов for, while. Пользуйтесь возможностями numpy.** 

```python
import numpy as np
```

```python
def euclidian_metric(X, x):
    distances = np.sqrt(np.sum((X-x) ** 2, axis = 1))
    return distances

def manhattan_metric(X, x):
    distances = np.sum(np.abs((X-x)), axis = 1)
    return distances
```

### Реализуйте алгоритм kNN для регрессии (4 балла)

- реализуйте класс kNN для равномерных весов (то есть $w_1 = \ldots = w_k$) **(3 балла)**
- добавьте возможность передать данному классу параметр `weights='distance'` для вызова взвешенной версии алгоритма kNN (то есть $w_i = \frac{1}{d\left(x, x_{(i)}\right)}$ **(1 балл)**

```python
class KNN:
    def __init__(self, metric='euclid', k=5, weights='uniform'):
        """
        PARAMETERS:
        metric ('euclid' or 'manhattan')
        k - number of nearest neighbors
        """

        self.metric = metric
        self.k = k
        self.weights = weights
        
        if metric == 'euclid':
            self.eval = euclidian_metric
        else:
            self.eval = manhattan_metric
        
        self.X_train = None
        self.y_train = None
        
    def fit(self, X_train, y_train):
        """
        INPUT:
        X_train - np.array of shape (n, d)
        y_train - np.array of shape (n,)
        """
        self.X_train = X_train
        self.y_train = y_train
        
        
    def predict(self, X_test):
        """
        INPUT:
        X_test - np.array of shape (m, d)
        
        OUTPUT:
        y_pred - np.array of shape (m,)
        """
        y_pred = []
        
        if self.weights == 'uniform':
        
            for observation in X_test:
                dist = self.eval(self.X_train, observation)
                select = dist <= np.sort(dist)[self.k - 1]
                y = np.mean(y_train[select])
                y_pred.append(y)
        
        else:
            for observation in X_test:
                dist = self.eval(self.X_train, observation)
                select = dist <= np.sort(dist)[self.k - 1]
                y = np.sum(y_train[select] * (1 / dist[select])) / np.sum(1 / dist[select])
                y_pred.append(y)

        y_pred = np.array(y_pred)
        
        return y_pred
```

# Линейная регрессия своими руками (5)

Реализуйте линейную регрессию с градиентным спуском для [функции потерь Хьюбера](https://en.wikipedia.org/wiki/Huber_loss):

$$
L_\delta\left(y, \hat{y}\right) =
\begin{cases}
\frac{1}{2}\left(y - \hat{y}\right)^2, \qquad &|y - \hat{y}| \leq \delta\\
\delta\left|y - \hat{y}\right| - \frac{1}{2}\delta^2,\qquad & \text{otherwise}
\end{cases}
$$

В таком случае общее значение функции потерь на всем датасете $(x_1, y_1), \ldots, (x_\ell, y_\ell)$ будет равно

$$
L = \frac{1}{\ell}\sum\limits_{i=1}^\ell L_\delta\left(y_i, \hat{y}_i\right)
$$

*Чему будет равен градиент этой функции по $w$? Вспомните, что за вектор $\hat{y}$ и как он зависит от $X$ и $w$.*

- https://stats.stackexchange.com/questions/312737/mean-absolute-error-mae-derivative

Учитывая, что $\hat y = Xw$, можем найти градиент $L_\delta\left(y, \hat{y}\right)$ по $w$:

$$
\nabla \left( L_\delta\left(y, \hat{y}\right) \right)_w =
\begin{cases}
-X \cdot (y - Xw), \qquad &|y - Xw| \leq \delta \\
-\delta X \cdot \text{sign}\, (y - Xw),\qquad & \text{otherwise}
\end{cases}
$$

*Примечание:* $y$, $w$ -- векторы; $X$ -- матрица. Для произвольного вектора $z = (z_1 \, \cdots \, z_\ell )$ справедливо $\text{sign} \, z =  (\text{sign} \, z_1 \, \cdots \, \text{sign} \, z_\ell )$. ### Реализуйте функцию потерь Хьюбера для одного примера и ее градиент по весам (1.5 балла)

- функция потерь **(0.5 балла)**
- градиент **(1 балл)**

**В данном пункте запрещено использование циклов for, while. Пользуйтесь возможностями numpy. Однако можно использовать оператор if.**

```python
def huber_loss(x, y, w, delta):
    """
    x - np.array shape=(d,)
    y - scalar
    w - np.array shape=(d,)
    delta - scalar
    
    OUTPUT:
    loss - scalar
    """
    
    if np.abs(y - np.dot(x, w)) <= delta:
        return 0.5 * (y - np.dot(x, w)) ** 2
    
    else:
        return delta * np.abs(y - np.dot(x, w)) - 0.5 * delta **2
    
    pass

def huber_grad(x, y, w, delta):
    """
    INPUT:
    x - np.array shape=(d,)
    y - scalar
    w - np.array shape=(d,)
    delta - scalar
    
    OUTPUT:
    grad - np.array shape=(d,)
    """

    if np.abs(y - np.dot(x, w)) <= delta:
        return - x * (y - np.dot(x, w))
        
    elif y - np.dot(x, w) > 0:
        return - delta * x
        
    else:
        return delta * x
        
    pass
```

```python
# проверка

x = np.array([1, 2, 3])
w = np.array([3, 5, 12])
y = 19
delta = 1

assert huber_loss(x, y, w, delta) == 29.5
assert np.allclose(huber_grad(x, y, w, delta), np.array([1, 2, 3]))

y = 49.2

assert np.allclose(huber_loss(x, y, w, delta), 0.02000000000000057)
assert np.allclose(huber_grad(x, y, w, delta), np.array([-0.2, -0.4, -0.6]))
```

### Реализуйте линейную регрессию (3.5 балла)

*Вы можете опустить единичный признак в модели и не добавлять его в данные. Для данной искусственной выборки это не актуально, потому что целевая переменная в этом случае является случайной величиной из стандартного нормального распределения со средним 0.*

*Вектор весов в градиентном спуске можете инициализировать нулями.*

```python
from numpy import linalg as LA
```

```python
class LinearRegressionHuber:
    def __init__(self, delta=1.0, max_iter=1000, tol=1e-6, eta=1e-2):
        """
        PARAMETERS:
        delta - scalar in Huber loss
        max_iter - maximum possible number of iterations in Gradient Descent
        tol - precision for stopping criterion in Gradient Descent
        eta - step size in Gradient Descent (learning rate)
        """

        self.delta = delta
        self.max_iter = max_iter
        self.tol = tol
        self.eta = eta
        
        self.w = None
        self.loss_history = None
        
    def fit(self, X, y):
        """
        INPUT:
        X_train - np.array of shape (n, d)
        y_train - np.array of shape (n,)
        
        В этой функции вы должны инициализировать веса (можно нулями), а также 
        итерационно обновлять веса с помощью 
        градиентного спуска (считать и запоминать лосс (значение функции потерь) будет хорошим решением)
        """

        self.X_train = X
        self.y_train = y
        
        self.w = np.zeros(self.X_train.shape[1])
        self.loss_history = []
        
        while len(self.loss_history) < self.max_iter:
            self.w -= self.eta * self.calc_gradient(self.X_train, self.y_train)
            self.loss_history.append(self.calc_loss(self.X_train, self.y_train))
            if LA.norm(self.eta * self.calc_gradient(self.X_train, self.y_train)) < self.tol:
                break
        
        self.loss_history = np.array(self.loss_history)
        return self.loss_history
        
    def predict(self, X):
        """
        INPUT:
        X_test - np.array of shape (m, d)
        
        OUTPUT:
        y_pred - np.array of shape (m,)
        
        Предскажите ответы с помощью обученных весов
        """
        
        y_pred = np.dot(X, self.w)

        return y_pred
    
    def calc_gradient(self, X, y):
        """
        Calculates the gradient of Huber loss by weights.
        
        INPUT:
        X - np.array of shape (l, d)
        y - np.array of shape (l,)
        
        OUTPUT:
        grad - np.array of shape (d,)
        
        Посчитайте градиент как среднее от градиентов для каждого примера
        """

        grad = np.zeros_like(self.w)

        grads = [huber_grad(observation, answer, self.w, self.delta) for observation, answer in zip(self.X_train, self.y_train)]
        grad = np.mean(grads, axis = 0)
        
        return grad 
    
    def calc_loss(self, X, y):
        """
        Calculates the Huber loss.
        
        INPUT:
        X - np.array of shape (l, d)
        y - np.array of shape (l,)
        
        OUTPUT:
        loss - float
        
        Посчитайте loss по выборке как среднее loss'ов для каждого 
        примера
        """

        loss = 0
        
        losses = [huber_loss(observation, answer, self.w, self.delta) for observation, answer in zip(X, y)]
        loss = np.mean(losses)
        
        
        return loss
```

## 1* (0.5)

Найдите константу $C$, решающую следующую задачу ($0 < \tau < 1$ фиксировано):

$$
\sum\limits_{i=1}^\ell\rho_\tau\left(y_i - C\right) \to \min_C,
$$

где

$$
\rho_\tau(z) =
\begin{cases}
\tau z, & z > 0\\
(\tau - 1)z, & z \leq 0
\end{cases}
$$

#### Решение

<font color='darkred'> Ход решения этой задачи приведен после решения задачи №2. **Спойлер:** C = $\tau$-квантиль выборки. </font>

## 2* (0.5)

Алгоритм регрессии выдает константный ответ для любого объекта: $a(x) = C$. Найдите $C$, при котором средняя абсолютная ошибка на обучающей выборке $\{(x_1, y_1), \ldots, (x_\ell, y_\ell)\}$ будет минимальной.

#### Решение

*Наша задача сводится к поиску аргумента минимума функции MAE (Mean Absolute Error):*

$$ 
   \arg \min_{C} \frac{1}{\ell} \sum_{i=1}^{\ell} \left| y_i - C \right| \\
   \arg \min_{C} \sum_{i=1}^{\ell} \left| y_i - C \right|
$$

Предположим, что наблюдения упорядочены по целевой переменной, то есть $y_1 \leq y_2 \leq \ldots \leq y_\ell$. Заметим, что если $C \leq y_k$, то справедливо равенство:

$$
    \sum_{i=k}^{\ell} \left| y_i - C \right| = \sum_{i=k}^{\ell} \left( y_i - C \right)
$$

Аналогично, если $C \geq y_k$, то   $\sum_{i=1}^{k} \left| y_i - C \right| = \sum_{i=1}^{k} \left( C - y_i \right)$. Теперь можем двигаться дальше.

Пусть $y_k \leq C \leq C + \Delta C \leq y_{k+1}$. Тогда:

$$
    \text{MAE}(C + \Delta C) = \\[1ex]
    = \sum_{i=1}^{k} \left( C + \Delta C - y_i \right) + \sum_{i=k+1}^{\ell} \left( y_i - (C + \Delta C) \right) = \\[1ex]
    = \Delta C \cdot k + \sum_{i=1}^{k} \left( C - y_i \right) - \Delta C \cdot (\ell - k) + \sum_{i=k+1}^{\ell} \left( y_i - C  \right) = \\[1ex]
    \Delta C \cdot (2k - \ell) + \sum_{i=1}^{k} \left( C - y_i \right) + \sum_{i=k+1}^{\ell} \left( y_i - C  \right) = \\[1ex]
    = \Delta C \cdot (2k - \ell) + \text{MAE} (C)
$$

Получаем, что $\text{MAE}(C + \Delta C) = \Delta C \cdot (2k - \ell) + \text{MAE} (C) $, то есть 

$$\underbrace{\text{MAE}(C + \Delta C) - \text{MAE} (C)}_{\Delta \text{MAE}} = \Delta C \cdot (2k - \ell) $$

В таком случае, $\text{MAE}$ возрастает (убывает) при $2k > \ell $ ($2k < \ell$) и остается неизменной при $2k = \ell$. Если минимум $\text{MAE}$ существует, то он достигается при $k = \frac{\ell}{2}$, а значит

$$
    y_{\frac{\ell}{2}} \leq C \leq y_{\frac{\ell}{2} + 1}
$$

Это и говорит о том, что $C_{\text{min}} = M (y_1, \ldots, y_\ell)$, т. к. $C$ делит упорядоченную выборку так, что одна половина наблюдений оказывается меньше $C$, другая -- больше $C$.

$$
    \boxed{C = M (y_1, \ldots, y_\ell)}
$$

#### Решение задачи №1

Пусть $\displaystyle L = \sum\limits_{i=1}^\ell\rho_\tau\left(y_i - C\right) \to \min_C $. Аналогично задаче №2, постулируем два равенства для упорядоченной выборки:

$$
    L = 
    \begin{cases} 
       \displaystyle \sum_{i=k}^{\ell} \tau (y_i - C), \quad & C \leq y_k \\
       \displaystyle \sum_{i=1}^{k} \tau (y_i - C), \quad & C \geq y_k
    \end{cases}
$$

Пусть $y_k \leq C \leq C + \Delta C \leq y_{k+1}$. Тогда:

$$
    L(C + \Delta C) = \\[1ex]
    \sum_{i=1}^{k}(\tau - 1)(y_i - (C + \Delta C)) + \sum_{i = k + 1}^{\ell} \tau (y_i - (C + \Delta C)) = \\[1ex]
    -k\Delta C(\tau - 1) + \sum_{i=1}^{k} (\tau - 1) (y_i - C) - \tau \Delta C (\ell - k) + \sum_{i = k + 1}^{\ell} \tau (y_i - C) = \\[1ex]
    \Delta C(k - \tau \ell) + L(C)
$$

Имеем $L(C + \Delta C) - L(C) = \Delta C(k - \tau \ell)$. Значит, минимум функции достигается при $k = \tau \ell$. Таким образом,

$$
    y_{\tau \ell} \leq C \leq y_{\tau \ell + 1}
$$

В таком случае, $C_{\text{min}}$ -- это квантиль порядка $\tau$.

$$
    \boxed{C = Q_{\tau} (y_1, \ldots, y_\ell)}
$$

## 3* (0.5)

Гарри Поттер хочет найти философский камень, расположенный в точке минимума функции $f(x_1, x_2)=x_1^2 + x_2^2$. В момент времени 0 он стартует из точки $x^{(0)}=(2, 2)$. На $i$-й минуте Гарри мгновенно перемещается (аппарирует) из точки $x^{(i)}$ в точку

$$
x^{(i+1)} = x^{(i)} - \eta \nabla f(x^{(i)}),
$$

где $\nabla f(x^{(i)})$ — градиент $f$ в точке $x^{(i)}$, $\eta \ge 0$ — фиксированное число. Опишите судьбу Гарри в зависимости от значения $\eta$. При каких значениях $\eta$ Гарри подойдёт к философскому камню сколь угодно близко? Сколько итераций ему понадобится, чтобы подойти к философскому камню на расстояние не больше $\varepsilon$?

#### Решение

Для начала найдем градиент функции $f$:

$$ \nabla f = \begin{pmatrix} 2x_1 \\ 2x_2 \end{pmatrix} $$

Сразу очевидно, что минимум функции находится в точке $(0, 0)$, так как вектор-градиент равен 0 только в этой точке.

Заметим, что, хотя задача и определена для двух переменных, она симметрична относительно их (градиент и начальное условие совпадают). Поэтому рассмотрим только задачу для $x_1$ (обозначим его как $x$) без потери общности.

$$
        x(i + 1) = x(i) - 2\eta x(i) \\
        x(0) = 2
$$

$x(i + 1) + (2\eta - 1) \cdot x(i) = 0(i)$ -- простейшее разностное уравнение первого порядка. Его решение имеет вид $x(i) = C \cdot (1 - 2\eta)^i$, где $C$ -- произвольная константа. Из начального условия 

$$ x(0) = 2 = C \quad \Rightarrow \quad x(i) = 2 \cdot (1 - 2\eta)^i$$

Полученное равенство и задает "судьбу" Гарри, позволяя вычислить его координаты на $i$-м шаге. Поскольку философский камень находится в нуле, можно проверить сходимость $x(i)$ к нулю при $i \to \infty$:

$$
        \lim_{i \to \infty}  x(i) = \lim_{i \to \infty} \left[ 2 \cdot (1 - 2\eta)^i  \right] = 0 \quad \Longleftrightarrow \quad -1 < 1 - 2\eta < 1 \\
        \boxed{0 < \eta < 1}
$$

При $\forall \, \eta \in (0; 1)$ Гарри Поттер подойдет к камню сколь угодно близко.

Расстояние от Гарри до камня задается как $\left|x(i) - 0\right| = \left|x(i)\right| = \left|2 \cdot (1 - 2\eta)^i\right|$. Решим неравенство

$$
        \left|2 \cdot (1 - 2\eta)^i\right| \leq \varepsilon \\[1ex]
        \left|(1 - 2\eta)^i\right| \leq \frac{\varepsilon}{2} \\[1ex]
        \left|1 - 2\eta\right|^i \leq \frac{\varepsilon}{2} \\[1ex]
        i \cdot \underbrace{\log \left|1 - 2\eta\right|}_{\leq 0 \, \text{if} \, \eta \in (0; 1)} \leq \log \frac{\varepsilon}{2} \\[1ex]
        i \geq \frac{\log \varepsilon - \log 2}{\log \left|1 - 2\eta\right|}
$$

Тогда Гарри понадобится $\left \lceil \frac{\log \varepsilon - \log 2}{\log \left|1 - 2\eta\right|} \right \rceil$ итераций.
